```python
def set_size(width = "thesis", fraction=1, subplots=(1, 1)):
    if width == 'thesis':
        width_pt = 426.79135
    elif width == 'beamer':
        width_pt = 307.28987
    else:
        width_pt = width

    # Width of figure (in pts)
    fig_width_pt = width_pt * fraction

    # Convert from pt to inches
    inches_per_pt = 1 / 72.27

    # Golden ratio to set aesthetic figure height
    golden_ratio = (5**0.5 - 1) / 2

    # Figure width in inches
    fig_width_in = fig_width_pt * inches_per_pt
    # Figure height in inches
    fig_height_in = fig_height_in = fig_width_in * golden_ratio * (subplots[0] / subplots[1])

    fig_dim = (fig_width_in, fig_height_in)

    return fig_dim
```

```python
ldist = LawrenceDist(2, 7)
```

```python
probs = ldist.calc_prob(4.43)
```

```python
plt.figure(figsize = set_size())
plt.bar(ldist.phase_space, probs)
plt.xlabel("Phase Space $\Omega$")
plt.ylabel("Probability")
plt.title("Lawrence' Distribution with $\\langle n \\rangle = 4.43$")
plt.tight_layout()
plt.savefig("../output/misc/lawrence_dist.pdf")
```

```python
plt.figure(figsize = set_size())
d = np.linspace(0, 1)
plt.plot(d, d * (1-d), label = "$d \\cdot (1 - d)$")
plt.xlabel("Decimals d")
plt.legend()
plt.ylabel("Variance")
plt.title("Variance of Lawrence' Distribution")
plt.tight_layout()
#plt.savefig("../output/misc/lawrence_dist_variance.pdf")
```
