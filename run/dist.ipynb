```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
loc = 20
scale = 8
```

```python
s = np.random.normal(loc, scale, size = (100000))
s1 = np.floor(s)
s2 = np.ceil(s)
```

```python
t = np.concatenate((s1, s2), axis = 0)
```

```python
plt.hist(t, bins = np.arange(-30,30))
```

```python
print("mean =", t.mean())
print("std =", t.std())
```
