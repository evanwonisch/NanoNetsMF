```python
import numpy as np
from matplotlib import pyplot as plt
from module.base.network import Network
from module.simulation.set_meanfield import SetMeanField
import module.components.CONST as CONST
```

```python
set = SetMeanField(0.05, 0)
set.n = np.array([3.])
set.n2 = 10
```

```python
ns = []
n2s = []

for i in range(100):
    set.evolve(N = 1, dt = 0.01)
    ns.append(np.copy(set.n))
    n2s.append(np.copy(set.n2))

ns = np.array(ns)
n2s = np.array(n2s)
```

```python
print("n =", set.n)
print("n2 =", set.n2)
print("var = ", set.n2 - set.n**2)
print("dn =", set.calc_dn_alt())
print("dn2 =", set.calc_dn2_alt())
```

```python
plt.plot(ns, label = "<n>")
plt.plot(np.array(ns) ** 2)
plt.plot(n2s, label = "<n^2>")
plt.legend()
```
